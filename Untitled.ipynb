```python
import pandas as pd
import glob
```

```python
FRAC = 1/1000
OUTPUT_PATH = '../sample_bike_data2.csv'
```

```python
csv_paths = glob.glob('*.csv') 
csv_paths
```

```python
df = pd.DataFrame()
```

```python
for path in csv_paths:
    #     print(path)
    new_df = pd.read_csv(path, low_memory=False)
    df = pd.concat([df, new_df])

df
```

```python
# 5845324-len(csv_paths)
```

```python
sample_df = df.sample(frac=FRAC, random_state=42) 
sample_df
```

```python
sample_df.to_csv(OUTPUT_PATH)
```
